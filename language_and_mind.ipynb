```python
from fpdf import FPDF
import os

class KoreanPDF(FPDF):
    def __init__(self):
        super().__init__()
        self.add_font('NanumGothic', '', '/Users/bagjuhyeon/Library/Fonts/KyoboHandwriting2019.otf', uni=True)
    
    def korean_text(self, x, y, txt):
        self.set_xy(x, y)
        self.write(h=5, txt=txt)

# PDF 인스턴스 생성
pdf = KoreanPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
pdf.set_font('NanumGothic', size=12)

# 제목
pdf.set_font('NanumGothic', size=16)
pdf.cell(200, 10, txt="문제 및 정답", ln=True, align="C")
pdf.ln(10)

# 내용
content = [
    ("1. 객관식 문제", "담화 처리를 통해 이루어지는 주요 과정이 아닌 것은 무엇입니까?",
     ["1) 텍스트 내 단어를 해당 개념이나 대상으로 연결하는 과정",
      "2) 텍스트가 무엇에 관한 것인지 내용 표상을 구축하는 과정",
      "3) 이야기 속 사건의 연속성을 이해하는 과정",
      "4) 텍스트 외부 정보와 연결을 차단하는 과정"], 
     "정답: 4) 텍스트 외부 정보와 연결을 차단하는 과정"),
    # ... 나머지 content 내용 ...
]

# PDF에 내용 추가
current_y = pdf.get_y()
for num, question, options, answer in content:
    # 문제 번호
    pdf.set_font('NanumGothic', size=12)
    pdf.set_xy(10, current_y)
    pdf.korean_text(10, current_y, num)
    current_y += 10
    
    # 문제
    pdf.korean_text(10, current_y, question)
    current_y += 10
    
    # 보기
    if options:
        for opt in options:
            pdf.korean_text(10, current_y, opt)
            current_y += 10
    
    current_y += 5
    
    # 정답
    pdf.korean_text(10, current_y, answer)
    current_y += 15

# PDF 저장
pdf.output("문제와_정답.pdf", 'F')
print("PDF 파일이 '문제와_정답.pdf' 이름으로 저장되었습니다.")
```
